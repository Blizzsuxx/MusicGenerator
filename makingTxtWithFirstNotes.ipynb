```python
def make_dictionary_first_notes(path):
    dictionary = {}
    num=0;
    for filename in listdir(path):
        print(filename)
        if num%25 != 0:
            num += 1
            continue
        num += 1
        file = open("all_of_the_data/"+filename, 'r') 
        text = file.read() 
        file.close()
        file2 = open("first_notes.txt", 'a')   
        first_notes = text.split(" ")[0:5]
        line = filename.split(".txt")[0] + " " +first_notes[0] + " " +first_notes[1] + " " +first_notes[2] + " " +first_notes[3] + " " +first_notes[4] +  "\n"
        print(line)
        
        file2.writelines(line)
        file2.close()
        file.close()
        
        dictionary[filename] = first_notes
    return dictionary
```

```python
make_dictionary_first_notes("all_of_the_data/")
```
