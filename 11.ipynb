```python
import os 
from os import listdir
```

```python
def rename(path):
    for filename in listdir(path):
        if len(filename.split(".mid"))>1:
            new_name = filename.split(".mid")[0] + filename.split(".mid")[1]
            os.rename("C:/soft/bb/MusicGenerator/data5/" +filename, new_name)
rename("C:/soft/bb/MusicGenerator/data5")
```

```python
def make_dictionary_first_notes(path):
    dictionary = {}
    num=0;
    for filename in listdir(path):
        if num >2200:
            break
        file = open("C:/soft/bb/MusicGenerator/data5/"+filename, 'r') 
        text = file.read() 
        file.close()
#         file = open("C:/soft/bb/MusicGenerator/first_notes.txt", 'a')   
        first_notes = text.split(" ")[0:5]
        line = filename.split(".txt")[0] + " " +first_notes[0] + " " +first_notes[1] + " " +first_notes[2] + " " +first_notes[3] + " " +first_notes[4] +  "\n"
        file.writelines(line)
        file.close()
        num += 1
        dictionary[filename] = first_notes
    return dictionary
```

```python
print(make_dictionary_first_notes("C:/soft/bb/MusicGenerator/data5"))
```
